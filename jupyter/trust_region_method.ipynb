# 信赖域方法
想法：在一局部区域（信赖域）用一模型函数逼近，求模型函数在信赖域中的最小值点

### 推导过程
$$ f(x_k +p) = f(x_k) + \nabla f(x_k)p + \frac{1}{2}p^T \nabla^2 f(x_k)p$$
近似模型
$$ m_k(p) = f + g^Tp + \frac{1}{2}p^TB_kp , |p| <\Delta  $$

### 判断准则
$$ \rho_k = \frac{f(x_k)-f(x_k +p_k)}{m_k(0) - m_k(p_k)}$$
若 $\rho_k \leq 0$, 则拒绝这一步，并且缩小$\Delta x_k$  

若 $\rho_k \approx 1$, 则接受这一步，并且下一步扩大$\Delta x_k$  

若 $\rho_k \approx 0$, 则拒绝这一步，并且缩小$\Delta x_k$  

### 算法过程
Given $\hat{\Delta} > 0, \Delta_0 \in (0,\hat{\Delta}),\eta \in (0,\frac{1}{4})$  
While $ g(x_k) > \epsilon $  
算出$p_k$与 $\rho_k$  
if $ \rho_k < \frac{1}{4}$  
    $\Delta_{k+1} = \frac{1}{4} \Delta_k$ 
  


### 难点 
求解$ m_k(p) = f + g^Tp + \frac{1}{2}p^TB_kp , |p| <\Delta  $的约束优化是困难的  
如果$\Delta$足够大，使得最优点在信赖域里面，那么此时是简单的  
### Cauchy Point
先进行线性近似，选取方向，再选取步长,使得模型函数最小
$$ p^s_k = arg \; \min_{|p| < \Delta} \; f + g^Tp $$
$$ \tau_k = arg \; \min_{|\tau p^s_k| < \Delta} \; m_k(\tau p^s_k) $$
可以解出来
$$p^s_k = \frac{\Delta}{\lVert \nabla f(x_k) \lVert} \nabla f(x_k) $$

### Dogleg Method

# 带约束优化
### 一般形式
$$ \min_{x\in R^n} f(x)$$
$$ s.t.\begin{cases} c_i(x) = 0 &i\in \varepsilon\\c_j(x) \ge 0 & j\in \mathcal{J}\end{cases}$$
## 等值约束
$\textbf{必要条件:}$ 
$$\{f\text{的可行方向}\} \cap \{f\text{的下降方向}\} = \emptyset $$
$$\nabla f(x) = \lambda \nabla c_i(x) $$
引入lagrange乘子：$$ L(x,\lambda_1) := f(x) - \lambda c_1(x)$$ 
则必要条件可以写为$$ \nabla_x L(x,\lambda_1) := \nabla f(x) - \lambda \nabla c_1(x) = 0$$
注意：$\lambda$的正负并不能表示此时是极大值还是极小值
## 不等式约束
$\textbf{必要条件:}$ 
$$ \nabla_x L(x,\lambda_1) := \nabla f(x) - \lambda \nabla c_1(x) = 0,\lambda \ge 0$$
并且
$$ \lambda c_1(x) = 0$$
### 活跃集
下标集合$\mathcal{A}(x)$称为点x的活跃集(也被称为积极集),是有全部等值约束的下
标、在x点满足$c_i(x)=0$的不等值约束的下标组成。即$$ \mathcal{A}(x) := \varepsilon \cup \{j \in \mathcal{J} | c_j(x) = 0 \}$$
### 线性化可行方向
在前面的分析中,我们在考虑是否存在可行的下降方向时,是用一阶泰勒展开分析的。
这需要假设一阶线性展开能够捕捉可行域的几何性质。如果在x附近,一阶泰勒展开式与可行域相差巨大,那么我们不可能期望一阶线性逼近能够很好地近似原问题。
因此我们需要对活跃的约束条件c;在x处的性质,保证线性逼近能够很好地近似真实的可行域。  
在给定一个点x和活跃集$\mathcal{A}(x)$,定义线性化方向集:
$$ \mathcal{F}(x) := \{d|d^T \nabla c_i(x) = 0, \forall i \in \varepsilon;d^T \nabla c_j(x）\ge 0 ,\forall j \in \mathcal{A} \cap \mathcal{J}\} $$
要使线性方向集近似为切锥：要满足限制  
线性无关约束限制(LICQ):给定一个可行点.$ x \in \omega$以及活跃集$\mathcal{A}(x)$,如果
活跃的约束条件的梯度$\nabla c_i(x),\forall i \in \mathcal{A}(x)$是线性无关的。
